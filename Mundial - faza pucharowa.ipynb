```python
from sklearn.linear_model import LogisticRegressionCV

clf = LogisticRegressionCV(cv=10, random_state=2022, max_iter=10000).fit(X_train, y_train)
predictions = clf.predict(X_test)
predictions_train = clf.predict(X_train)
print("Zbiór treningowy: \n", confusion_matrix(y_train,predictions_train))
print(classification_report(y_train,predictions_train))
print("Zbiór testowy: \n", confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```
